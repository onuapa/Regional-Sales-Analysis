# Reading Data for Analysis

## Importing Required Libraries

Let's start by importing the necessary libraries for data manipulation and analysis:


```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading in the files

fact = pd.read_excel("FactTable.xlsx")

sales_teams = pd.read_excel('Dim_SalesTeams.xlsx')

customers = pd.read_excel('Dim_Customers.xlsx')

products = pd.read_excel('Dim_Products.xlsx')

store_loc = pd.read_excel("Dim_Store location.xlsx")

regions = pd.read_excel("Dim_Regions.xlsx")


# Set the option to display all columns without truncation
pd.set_option('display.max_columns', None)
```

# Merging Data Files
As part of our regional sales analysis, we need to combine data from the files we have read in above.

```python
df = fact.merge(products,on='_ProductID').merge(customers,on='_CustomerID').merge(sales_teams,on='_SalesTeamID').merge(store_loc,on='_StoreID')
```

# Data Cleaning And Manipulation

Data cleaning and manipulation are essential steps to ensure that our data is accurate, consistent, and ready for analysis. In this section, we'll cover basic data cleaning tasks and demonstrate how to manipulate our data to create new insights.

## Removing Duplicates

We start by identifying and removing duplicate records, ensuring that each data point is unique:


```python
# Drop rows with any missing values
df = df.dropna()
```

## Adding New Columns
We can derive new insights by creating new columns based on existing data. For example, calculating Sales and Profit:

- Sales: Here we multiplied the `Order Quantity` column with the `Unit Price` column
- Profit: Here we Subtracted the `unit price` from the `unit cost` to get Profit.



```python
df['Sales'] = df['Order Quantity'] * df['Unit Price']

df.head()
```

```python
df['Profit'] = (df['Unit Price'] - df['Unit Cost']) * df['Order Quantity']

df.head()
```

## 	Sales Channel by the Quantity Ordered 

```python
df.head(23)
df['Sales Channel'].unique()
sales_channel=pd.DataFrame(df.groupby('Sales Channel')[['Order Quantity','Sales',"Profit"]].sum().reset_index())
sales_channel['Sales Percentage %']= round(sales_channel['Sales']/sales_channel['Sales'].sum()*100,2)
sales_channel['Profit Percentage %']=round(sales_channel['Profit']/sales_channel['Profit'].sum()*100,2)
sales_channel['Order Percentage %']=round(sales_channel['Order Quantity']/sales_channel['Order Quantity'].sum()*100,2)
sales_channel
```
